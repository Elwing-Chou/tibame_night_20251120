````python
# 99(平) 2004(閏) 2100(平) 2000(閏)
year = 99
if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            print("閏")
        else:
            print("平")
    else:
        print("閏")
else:
    print("平")
````

````python
year = 2100
if year % 400 == 0:
    print("閏")
elif year % 100 == 0:
    print("平")
elif year % 4 == 0:
    print("閏")
else:
    print("平")
````

````python
def is_leap_year(year):
    ans = False
    if year % 400 == 0:
        ans = True
    elif year % 100 == 0:
        ans = False
    elif year % 4 == 0:
        ans = True
    else:
        ans = False
    return ans
print(is_leap_year(2000))
````

````python
# 判斷日期是否valid
# 20240228: valid
# 20240229: valid
# 20250229: invalid
# 20251301: invalid
date_checked = 20251301
# 年: easy 月: easy 日: 清單 [31, 28, 31...]
# 年月日 分出來
# 20240228 / 100 = 202402...28
day = date_checked % 100
remain = date_checked // 100
month = remain % 100
year = remain // 100
print(year, month, day)
# valid/invalid印出來
if year > 0:
    if 1 <= month <= 12:
        day_limit_per_month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
        # 清單操作: [0123...]
        day_limit = day_limit_per_month[month-1]
        if month == 2 and is_leap_year(year) == True:
            day_limit = day_limit + 1
        if 1 <= day <= day_limit:
            print("日期合理")
        else:
            print("日期不合理")
    else:
        print("月份不合理")
else:
    print("年份不合理")
````
